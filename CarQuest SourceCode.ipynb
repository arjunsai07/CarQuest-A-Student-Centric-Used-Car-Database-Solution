```python
import random
import string

def generate_random_vin():
    # VINs are typically 17 characters long
    length = 17
    # Characters used in VINs (excluding I, O, and Q to avoid confusion with numbers)
    characters = string.ascii_uppercase.replace('I', '').replace('O', '').replace('Q', '') + string.digits
    return ''.join(random.choice(characters) for _ in range(length))
    
ent['VIN'] = ent['VIN'].apply(lambda x: generate_random_vin() if pd.isna(x) else x)
```

```python

duplicate_vins = ent[ent.duplicated('VIN', keep=False)]
print(len(duplicate_vins))
print(duplicate_vins)

def generate_unique_vin(vin, counter):
    return f"{vin}_{counter}"

for vin in duplicate_vins['VIN'].unique():
    # Filter rows with the same VIN
    rows_with_duplicate_vin = ent[ent['VIN'] == vin]

    # Generate new unique VINs for each duplicate, except the first one
    counter = 1
    for index in rows_with_duplicate_vin.index[1:]:
        ent.at[index, 'VIN'] = generate_random_vin()
        counter += 1
print(ent[ent.duplicated('VIN', keep=False)])  # This should now be empty


```
